## Activity Over Time

```python
data['question_sent_date'] = pd.to_datetime(data['question_sent'].astype(str).str[:10], errors='coerce')
data['response_sent_date'] = pd.to_datetime(data['response_sent'].astype(str).str[:10], errors='coerce')

fig, ax = plt.subplots(figsize=(10,5))
data['question_sent_date'].dt.date.value_counts().sort_index().plot(ax=ax)
ax.set_title('Question Volume Over Time')
ax.set_xlabel('Date')
ax.set_ylabel('Number of Questions')
plt.show()
```

```python
data['question_sent'] = pd.to_datetime(data['question_sent'], utc=True, errors='coerce')
     
data['response_sent'] = pd.to_datetime(data['response_sent'], utc=True, errors='coerce')

data['response_lag'] = (data['response_sent'] - data['question_sent']).dt.total_seconds() / 3600

sns.histplot(data['response_lag'], bins=50)
plt.title('Response Lag Distribution (hours)')
plt.show()
```

```python
# Ensure both columns are parsed as datetimes
data['response_user_birthdate'] = pd.to_datetime(data['response_user_dob'], errors='coerce')
data['response_user_logindate'] = pd.to_datetime(
    data['response_user_created_at'].astype(str).str[:10], errors='coerce'
)

# Compute preliminary age difference
age_years = data['response_user_logindate'].dt.year - data['response_user_birthdate'].dt.year

# Determine if the birthday has occurred this year
has_had_birthday = (
    (data['response_user_logindate'].dt.month > data['response_user_birthdate'].dt.month) |
    (
        (data['response_user_logindate'].dt.month == data['response_user_birthdate'].dt.month) &
        (data['response_user_logindate'].dt.day >= data['response_user_birthdate'].dt.day)
    )
)

# Subtract 1 if the birthday hasnâ€™t happened yet this year
data['response_user_age'] = age_years - (~has_had_birthday).astype(int)
```

```python
# Parse both columns properly
data['question_user_birthdate'] = pd.to_datetime(data['question_user_dob'], errors='coerce')
data['question_user_logindate'] = pd.to_datetime(
    data['question_user_created_at'].astype(str).str[:10], errors='coerce'
)

# Calculate preliminary age difference
age_years = (
    data['question_user_logindate'].dt.year - data['question_user_birthdate'].dt.year
)

# Check if birthday has already occurred this year
has_had_birthday = (
    (data['question_user_logindate'].dt.month > data['question_user_birthdate'].dt.month) |
    (
        (data['question_user_logindate'].dt.month == data['question_user_birthdate'].dt.month) &
        (data['question_user_logindate'].dt.day >= data['question_user_birthdate'].dt.day)
    )
)

# Adjust by subtracting 1 if birthday hasnâ€™t happened yet
data['question_user_age'] = age_years - (~has_had_birthday).astype(int)
```

```python
sns.histplot(data['question_user_age'], bins=20)
plt.title('Distribution of Question User Age')
plt.show()
```

```python
sns.histplot(data['response_user_age'], bins=20)
plt.title('Distribution of Response User Age')
plt.show()
```
